```python
import pandas as pd
```

```python
df1 = pd.read_csv('../data_cleaned/merged_nominated_company_data.csv')
```

```python
df2 = pd.read_csv('../data_cleaned/merged_oscar_nominee.csv')
```

```python
df3 = pd.read_csv('../data_cleaned/cleaned_merged_df.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
columns_to_drop = ['Unnamed: 0']
df2.drop(columns_to_drop, axis=1, inplace=True)
```

```python
columns_to_drop = ['Unnamed: 0', 'Unnamed: 0.1']
df3.drop(columns_to_drop, axis=1, inplace=True)
```

```python
df3.head()
```

```python
df3.isna().sum()
```

```python
df3['license'].fillna('unknown', inplace=True)
```

```python
result_1 = pd.merge(df2, df1[['title', 'company_nominated']], on='title', how='left')
```

```python
result_df = pd.merge(result_1, df3[['title', 'license']], on='title', how='left')
```

```python
result_df
```

```python
result_df.isna().sum()
```

```python
duplicates = result_df[result_df.duplicated(subset='title', keep=False)]
```

```python
duplicates
```

```python
result_df = result_df.drop_duplicates(subset='title', keep='first')
```

```python
def keep_first_word(genre_string):
    return genre_string.split(',')[0]
result_df['genres'] = result_df['genres'].apply(keep_first_word)
```

```python
result_df.to_csv('../data_cleaned/merged_data.csv')
```
